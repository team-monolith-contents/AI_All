**[문제2] 아래 코드를 실행 후 기울기/y절편을 조정해 보고, 어떠한 방식으로 평균 제곱 오차가 적어지도록 하였는지 이야기해보아요.** 

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import matplotlib.pyplot as plt
import numpy as np
import math

# loss 함수 입니다.
def loss(a,b, data_x,data_y):
    return sum(((a*data_x + b)- data_y)**2)/ len(data_x)

# ipywidget의 output을 그려주는 함수입니다.
def draw(fig,a,b):
    
    clear_output(wait=True) # ipywidget에 포함된 함수로 output을 초기화 시켜주는 함수입니다.
    margin = 5
    
    # 아래의 코드로 출력되는 결과는 모두 ipywidget output으로 보내줍니다.
    plt.scatter(data_x, data_y)
    plt.plot([min(data_x)-margin, max(data_x)+margin],
            [a*min(data_x)+b, a*max(data_x)+b], c = 'r')
    plt.xlim(0, 70)
    plt.ylim(0, 1200)
    plt.show()
    
    sign = "+" if b >= 0 else "-"
    print(f"{a} * x {sign} {abs(b)}")
    print("평균 제곱 오차 :", loss(a, b, data_x, data_y))
    print("[도전] 평균 제곱 오차를 2000 아래로 줄여보세요!")

# a 값이 변경될 때 실행될 콜백함수입니다.
def on_a_value_change(change):
    with out: # with out: block에 있는 코드들은 출력값을 ipywidget output으로 보내줍니다.
        draw(fig,change['new'], b.value)

# b 값이 변경될 때 실행될 콜백함수입니다.
def on_b_value_change(change):
    with out:
        draw(fig, a.value, change['new'])

a = widgets.IntSlider(description='기울기 :', min=0, max=30) # a 변수의 값을 조절할 수 있는 IntSlider를 만들어줍니다.
b = widgets.IntSlider(description='y절편 :', min=-300, max=300) # b 변수의 값을 조절할 수 있는 IntSlider를 만들어줍니다.
out = widgets.Output() # ipywidget의 output을 만들어줍니다.

data_x = np.array([12, 20, 30, 56])
data_y = np.array([150, 208, 500, 875])

display(a,b,out) # a,b,out을 출력단에 표시해줍니다.

with out:
    fig = plt.figure()
    draw(fig, a.value, b.value)

a.observe(on_a_value_change, names='value') # a가 변화할 때 마다 실행 될 콜백함수를 등록합니다.
b.observe(on_b_value_change, names='value') # b가 변화할 때 마다 실행 될 콜백함수를 등록합니다.
```

## **2. 선형회귀 실습**  


실습에서는 직장에서의 연차와 연봉에 대한 데이터셋을 분석하여,  
두 변수 간 상관관계를 선형회귀로 분석해보아요.  

https://www.kaggle.com/datasets/rohankayan/years-of-experience-and-salary-dataset/code?resource=download

```python
import numpy as np
import pandas as pd

df = pd.read_csv('Salary_Data.csv')
df.head(5)
```

```python
df.describe()
```

```python
x_data = np.array(df['YearsExperience'], dtype=np.float32).reshape((-1, 1)) 
y_data = np.array(df['Salary'], dtype=np.float32).reshape((-1, 1))

print(x_data.shape)
print(y_data.shape)
```

```python
import matplotlib.pyplot as plt

plt.scatter(x_data, y_data)
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.show()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_val, y_train, y_val = train_test_split(x_data, y_data, test_size=0.4, random_state=2021)
print(x_train.shape, x_val.shape)
print(y_train.shape, y_val.shape)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression(fit_intercept=True)
model.fit(x_train, y_train)
```

```python
print('y절편 : ', model.intercept_)
print('가중치 : ', model.coef_)
```

```python
import matplotlib.pyplot as plt

y_val = model.predict(x_val)

plt.scatter(x_val, y_val)
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.plot(x_val, y_val, color="red")
plt.show()
```

```python
import matplotlib.pyplot as plt

y = model.predict(x_data)

plt.scatter(x_data, y_data)
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.plot(x_data, y, color="red")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

print("평균 제곱 오차: " + str(mean_squared_error(model.predict(x_val), y_val)))
```

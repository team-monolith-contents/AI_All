$$\frac{74^2 + 112^2 + (-60)^2 + (-123)^2}{4} = 9187.25$$

ÏÑ†ÌòïÌöåÍ∑ÄÎäî Ïù¥ Í≥ºÏ†ïÏùÑ ÏàòÏ∞®Î°Ä Î∞òÎ≥µÌïòÏó¨, ÏµúÏ¢ÖÏ†ÅÏúºÎ°ú ÌèâÍ∑† Ï†úÍ≥± Ïò§Ï∞®Í∞Ä Í∞ÄÏû• Ï†ÅÏùÄ ÏÑ†ÏùÑ ÏÑ†ÌÉùÌïòÎäî Í≤ÉÏù¥ÏóêÏöî.



**[Î¨∏Ï†ú2] ÏïÑÎûò ÏΩîÎìúÎ•º Ïã§Ìñâ ÌõÑ Í∏∞Ïö∏Í∏∞/yÏ†àÌé∏ÏùÑ Ï°∞Ï†ïÌï¥ Î≥¥Í≥†, Ïñ¥Îñ†Ìïú Î∞©ÏãùÏúºÎ°ú ÌèâÍ∑† Ï†úÍ≥± Ïò§Ï∞®Í∞Ä Ï†ÅÏñ¥ÏßÄÎèÑÎ°ù ÌïòÏòÄÎäîÏßÄ Ïù¥ÏïºÍ∏∞Ìï¥Î≥¥ÏïÑÏöî.** 

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import matplotlib.pyplot as plt
import numpy as np
import math

# loss Ìï®Ïàò ÏûÖÎãàÎã§.
def loss(a,b, data_x,data_y):
    return sum(((a*data_x + b)- data_y)**2)/ len(data_x)

# ipywidgetÏùò outputÏùÑ Í∑∏Î†§Ï£ºÎäî Ìï®ÏàòÏûÖÎãàÎã§.
def draw(fig,a,b):
    
    clear_output(wait=True) # ipywidgetÏóê Ìè¨Ìï®Îêú Ìï®ÏàòÎ°ú outputÏùÑ Ï¥àÍ∏∞Ìôî ÏãúÏºúÏ£ºÎäî Ìï®ÏàòÏûÖÎãàÎã§.
    margin = 5
    
    # ÏïÑÎûòÏùò ÏΩîÎìúÎ°ú Ï∂úÎ†•ÎêòÎäî Í≤∞Í≥ºÎäî Î™®Îëê ipywidget outputÏúºÎ°ú Î≥¥ÎÇ¥Ï§çÎãàÎã§.
    plt.scatter(data_x, data_y)
    plt.plot([min(data_x)-margin, max(data_x)+margin],
            [a*min(data_x)+b, a*max(data_x)+b], c = 'r')
    plt.xlim(0, 70)
    plt.ylim(0, 1200)
    plt.show()
    
    sign = "+" if b >= 0 else "-"
    print(f"{a} * x {sign} {abs(b)}")
    print("ÌèâÍ∑† Ï†úÍ≥± Ïò§Ï∞® :", loss(a, b, data_x, data_y))
    print("[ÎèÑÏ†Ñ] ÌèâÍ∑† Ï†úÍ≥± Ïò§Ï∞®Î•º 2000 ÏïÑÎûòÎ°ú Ï§ÑÏó¨Î≥¥ÏÑ∏Ïöî!")

# a Í∞íÏù¥ Î≥ÄÍ≤ΩÎê† Îïå Ïã§ÌñâÎê† ÏΩúÎ∞±Ìï®ÏàòÏûÖÎãàÎã§.
def on_a_value_change(change):
    with out: # with out: blockÏóê ÏûàÎäî ÏΩîÎìúÎì§ÏùÄ Ï∂úÎ†•Í∞íÏùÑ ipywidget outputÏúºÎ°ú Î≥¥ÎÇ¥Ï§çÎãàÎã§.
        draw(fig,change['new'], b.value)

# b Í∞íÏù¥ Î≥ÄÍ≤ΩÎê† Îïå Ïã§ÌñâÎê† ÏΩúÎ∞±Ìï®ÏàòÏûÖÎãàÎã§.
def on_b_value_change(change):
    with out:
        draw(fig, a.value, change['new'])

a = widgets.IntSlider(description='Í∏∞Ïö∏Í∏∞ :', min=0, max=30) # a Î≥ÄÏàòÏùò Í∞íÏùÑ Ï°∞Ï†àÌï† Ïàò ÏûàÎäî IntSliderÎ•º ÎßåÎì§Ïñ¥Ï§çÎãàÎã§.
b = widgets.IntSlider(description='yÏ†àÌé∏ :', min=-300, max=300) # b Î≥ÄÏàòÏùò Í∞íÏùÑ Ï°∞Ï†àÌï† Ïàò ÏûàÎäî IntSliderÎ•º ÎßåÎì§Ïñ¥Ï§çÎãàÎã§.
out = widgets.Output() # ipywidgetÏùò outputÏùÑ ÎßåÎì§Ïñ¥Ï§çÎãàÎã§.

data_x = np.array([12, 20, 30, 56])
data_y = np.array([150, 208, 500, 875])

display(a,b,out) # a,b,outÏùÑ Ï∂úÎ†•Îã®Ïóê ÌëúÏãúÌï¥Ï§çÎãàÎã§.

with out:
    fig = plt.figure()
    draw(fig, a.value, b.value)

a.observe(on_a_value_change, names='value') # aÍ∞Ä Î≥ÄÌôîÌï† Îïå ÎßàÎã§ Ïã§Ìñâ Îê† ÏΩúÎ∞±Ìï®ÏàòÎ•º Îì±Î°ùÌï©ÎãàÎã§.
b.observe(on_b_value_change, names='value') # bÍ∞Ä Î≥ÄÌôîÌï† Îïå ÎßàÎã§ Ïã§Ìñâ Îê† ÏΩúÎ∞±Ìï®ÏàòÎ•º Îì±Î°ùÌï©ÎãàÎã§.
```

## **2. ÏÑ†ÌòïÌöåÍ∑Ä Ïã§Ïäµ**  


Ïã§ÏäµÏóêÏÑúÎäî ÏßÅÏû•ÏóêÏÑúÏùò Ïó∞Ï∞®ÏôÄ Ïó∞Î¥âÏóê ÎåÄÌïú Îç∞Ïù¥ÌÑ∞ÏÖãÏùÑ Î∂ÑÏÑùÌïòÏó¨,  
Îëê Î≥ÄÏàò Í∞Ñ ÏÉÅÍ¥ÄÍ¥ÄÍ≥ÑÎ•º ÏÑ†ÌòïÌöåÍ∑ÄÎ°ú Î∂ÑÏÑùÌï¥Î≥¥ÏïÑÏöî.  

https://www.kaggle.com/datasets/rohankayan/years-of-experience-and-salary-dataset/code?resource=download

```python
import numpy as np
import pandas as pd

df = pd.read_csv('Salary_Data.csv')
df.head(5)
```

```python
df.describe()
```

```python
x_data = np.array(df['YearsExperience'], dtype=np.float32).reshape((-1, 1)) 
y_data = np.array(df['Salary'], dtype=np.float32).reshape((-1, 1))

print(x_data.shape)
print(y_data.shape)
```

```python
import matplotlib.pyplot as plt

plt.scatter(x_data, y_data)
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.show()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_val, y_train, y_val = train_test_split(x_data, y_data, test_size=0.4, random_state=2021)
print(x_train.shape, x_val.shape)
print(y_train.shape, y_val.shape)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression(fit_intercept=True)
model.fit(x_train, y_train)
```

```python
print('yÏ†àÌé∏ : ', model.intercept_)
print('Í∞ÄÏ§ëÏπò : ', model.coef_)
```

```python
import matplotlib.pyplot as plt

y_val = model.predict(x_val)

plt.scatter(x_val, y_val)
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.plot(x_val, y_val, color="red")
plt.show()
```

```python
import matplotlib.pyplot as plt

y = model.predict(x_data)

plt.scatter(x_data, y_data)
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.plot(x_data, y, color="red")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

print("ÌèâÍ∑† Ï†úÍ≥± Ïò§Ï∞®: " + str(mean_squared_error(model.predict(x_val), y_val)))
```

## **1. CNN**  


### **(1) CNN의 등장 배경**  


> DNN만을 사용하였을 경우 이미지 처리에서 한계  
> 2차원 정보인데 1차원적으로 해석하다 보니 놓치는 정보 많음  
> CNN은 2차원에서 그림의 특징을 최대한 유지하는 모델  

> 이미지넷 이미지인식 경진대회  
> 2012년 부터 CNN 기반 AlexNet 우승  
> 현재는 사람의 능력 이하로 줄임  
> 그림: 역대 우승 알고리즘과 에러율  

> CNN을 사용하여 이미지 관련 다양한 task 처리 가능  
> object detection, semantic segmentation, image generation 등

### **(2) CNN의 구성 요소**  


> CNN에서는 두 가지 연산을 주로 사용  
> 1. convolution: 이미지와 필터가 있을 때 쿵짝쿵짝  
> 2. pooling: 이미지 크기 줄이는 작업  
>  
> convolution과 pooling 그림

> convolution할 때 필터에서 특정 이미지 특징을 잡아냄
> 다양한 이미지 특징을 잡기 위해서는 여러 개의 필터가 필요함
> * ex. 이미지의 가장자리를 추출하는 sobel filter
> * https://blog.naver.com/junghs1040/222220205229  
>
> convolution에서 filter의 개수를 channel이라고 함
> CNN의 전체 구조는 convolution과 pooling을 조합하여 사용
> alexnet, vgg 구조 그림

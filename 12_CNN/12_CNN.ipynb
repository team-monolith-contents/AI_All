### **(2) CNN의 구성 요소**  


CNN에서는 convolution과 pooling 두 가지 연산을 반복하여 사용해요.

**2. 풀링(pooling)**  
풀링 계층에서는 이미지의 크기를 줄일 수 있어요.  
기존의 크기 그대로 연산을 수행하면 오래 걸리기 때문에, 크기를 줄이면서도 중요한 특징은 남기도록 할 수 있어요.  
가장 대표적인 풀링의 예시는 Max(최댓값) pooling, Avg(평균) pooling이 있어요.

## **2. CNN 실습**  


이번 실습에서는 CNN과 관련한 대표적인 task 중 하나인 object detection을 체험해볼 거에요.  
원래 object detection을 하려면 오랜 시간 모델을 학습해야 해요.  
하지만 "Yolov3"라는 모델에서 학습된 네트워크를 제공해주므로, 이것을 다운로드하여 학습 없이도 사용할 수 있어요.  
여러 이미지에 대해 object detection을 사용해보고, 성능이 좋은지 살펴보아요.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Yolo 로드
net = cv2.dnn.readNet("yolov3.weights", "yolov3.cfg")

classes = []
with open("coco.names", "r") as f:
    classes = [line.strip() for line in f.readlines()]

layer_names = net.getLayerNames()
output_layers = [layer_names[i - 1] for i in net.getUnconnectedOutLayers()]
# output_layers = [layer_names[i[0] - 1] for i in net.getUnconnectedOutLayers()] # opencv 버전에 따라
colors = np.random.uniform(0, 255, size=(len(classes), 3))
```

```python
def predict_yolo(img_path):
  # 이미지 가져오기
  img = cv2.imread(img_path)
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  height, width, channels = img.shape

  blob = cv2.dnn.blobFromImage(img, 1/256, (416, 416), (0, 0, 0), swapRB=True, crop=False)
  net.setInput(blob) 
  outs = net.forward(output_layers)

  class_ids = []
  confidences = []
  boxes = []
  for out in outs:
      for detection in out:
          scores = detection[5:]
          class_id = np.argmax(scores)
          confidence = scores[class_id]
          if confidence > 0.5:
              # Object detected
              center_x = int(detection[0] * width)
              center_y = int(detection[1] * height)
              w = int(detection[2] * width)
              h = int(detection[3] * height)
              # 좌표
              x = int(center_x - w / 2)
              y = int(center_y - h / 2)
              boxes.append([x, y, w, h])
              confidences.append(float(confidence))
              class_ids.append(class_id)

  indexes = cv2.dnn.NMSBoxes(boxes, confidences, 0.5, 0.4)

  font = cv2.FONT_HERSHEY_PLAIN
  colors = np.random.uniform(0, 255, size=(len(boxes), 3))
  if len(indexes) > 0:
    for i in indexes.flatten():
        x, y, w, h = boxes[i]
        # print(x, y, w, h)
        label = str(classes[class_ids[i]])
        confidence = str(round(confidences[i], 2))
        color = colors[i]
        cv2.rectangle(img, (x, y), ((x+w), (y+h)), color, 3)
        cv2.putText(img, label + " " + confidence, (x, y+20), font, 5, (0, 255, 0), 3)

    plt.imshow(img)
  
  else:
    print('탐지된 물체가 없습니다.')
```

```python
import glob

paths = glob.glob('samples.jpg')
predict_yolo(paths[0])
```

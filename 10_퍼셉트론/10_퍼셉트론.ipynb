### **(2) 퍼셉트론 학습 원리**  


> 퍼셉트론의 구성 요소 - 가중치, bias  
> 전체 합이 임계치를 넘으면 1/아니면 0(step function)  
> 퍼셉트론은 그래프에서 직선으로 나타낼 수 있음  

> 퍼셉트론 loss 사용하여 loss를 줄이도록 함
> $$J(w)=\displaystyle\sum_{x_k \in Y}^{}{-y_k(w^Tx_k)}$$
> 실제값과 예측값 다르게 분류했다면 loss가 커짐

> 이터레이션 거치면서 각 경우 loss 계산하고 비교

> 비선형 공간에 대한 문제 -> **다층 퍼셉트론**  
> 활성화 함수를 사용하여 비선형 분류  
> 다층 퍼셉트론을 사용하여 여러 직선 + 비선형 기준으로 분류  
> 이것이 더욱 발전된 것이 Deep Neural Network, 자세한 것은 DNN에서 배움

<h2 style='font-weight: bold'>1. 인간의 뇌는 어떻게 작동할까?</h2>
</div>

인간의 뇌는 정말 복잡하고 놀라운 기관이에요. 이 작은 기관은 우리가 생각하고 움직이는 데에 아주 중요한 역할을 담당하고 있어요.  
아마도 여러분은 심장이 어떻게 뛰고 근육이 움직이며 생각할 수 있는지에 대해 궁금한 적이 있으실 거예요. 그럼 이제 한 번 뇌가 어떻게 동작하는지에 대해 알아볼까요?  
뇌는 많은 수의 신경세포로 이루어져 있어요. 이 신경세포들은 서로 연결되어 있고, 전기적인 신호를 주고받으며 작동하고 있답니다.

그림 1은 신경세포의 모습을 그린 것이에요. 뇌에는 이런 신경세포 1000억개가 네트워크 구성하고 있어요. 이 신경세포들은 다른 신경세포와 연결돼 있고, 이 연결은 작은 간극인 시냅스를 통해서 이루어져요. 신경세포는 전기적인 신호를 만들어내고, 이 신호는 축삭돌기를 통해 전달되어요. 축삭돌기는 다른 신경세포의 수상돌기와 연결돼 있으며, 시냅스를 통해 신호가 전달돼요.

뉴런에서 만들어진 신호는 뉴로전달물질이라는 전기적인 화학 물질을 이용해 시냅스를 통해 다른 신경세포로 전달돼요.  
이 신호 전달은 시냅스의 강도에 따라 조절되며, 신경세포들의 연결과 시냅스의 강도가 뇌의 학습과 기억 형성에 영향을 미치게 돼요.

뇌는 신경세포들 사이의 복잡한 상호작용을 통해 동작하며, 이를 통해 우리는 심장의 박동, 근육의 움직임, 인식, 사고, 감정 등을 경험하고 수행할 수 있어요. 하지만 아직까지 뇌의 작동 원리를 완전히 이해하는 것은 어려운 과제이며, 뇌의 복잡성과 다양성으로 인해 연구가 계속되고 있어요.

<h2 style='font-weight: bold'>2. 인공 신경망이란?</h2>
</div>

1957년 코넬 항공 연구소에 근무하던 신경과학자 프랭크 로젠블랫 (Frank Rosenblatt)은 이러한 뉴런의 구조를 본 따 퍼셉트론(Perceptron)을 고안해냈어요.  
인공 뉴런인 퍼셉트론을 네트워트로 연결하면 인간의 뇌처럼 판단과 결정을 수행하는 인공 신경망을 만들 수 있어요.  
아래 그림2와 그림3은 각각 신경망과 인공 신경망의 모습이에요. 그림3의 인공신경망의 파란 동그라미들이 뉴런의 역할을 하는 퍼셉트론이에요.  
여러분이 보기에는 인공 신경망과 실제 신경망의 모습이 닮은 것 같나요?

<h2 style='font-weight: bold'>3. 단층 퍼셉트론의 구조를 이해해보자.</h2>
</div>

그럼 퍼셉트론의 구조를 좀 더 자세하게 살펴보도록 할게요.




퍼셉트론은 입력층과 출력층으로 구성되어 있어요. 입력층은 초기 데이터를 받아들이는 역할을 하고, 출력층은 최종 결과를 내보내는 역할을 합니다. 입력층과 출력층 사이에는 가중치라는 값들로 연결된 노드들이 존재합니다.  
그림 5는 입력이 2개인 퍼셉트론을 나타낸 그림이에요. 왼쪽의 x1,x2가 각각 입력이고 오른쪽의 y가 출력값이에요. 가운데에 있는 w1과 w2는 가중치라는 부르는 값으로 인공 신경망의 퍼셉트론들은 이 가중치로 연결되어 있어요.  
각 퍼셉트론은 입력값과 가중치를 곱한 후 이를 합산하는 연산을 해요.  
그림 5의 예시에서는 입력 x1과 출력 y를 이어주는 가중치가 w1, 입력 x2와 출력 y를 이어주는 가중치가 w2 이므로 출력값 y를 계산하기 위해서는 다음과 같은 식을 사용할 수 있어요.

$$y = x_1*w_1 + x_2*w_2$$

퍼셉트론은 입력값과 가중치, 활성화 함수를 통해 학습하고 판단할 수 있게 됩니다.  
학습 과정에서는 입력값과 기대 출력값을 비교하여 가중치를 조정하게 되는데, 이를 통해 퍼셉트론은 패턴을 학습하고 새로운 데이터에 대해 예측이나 분류를 수행할 수 있게 됩니다.  
이렇게 입력층과 출력층 두 단계로만 이루어 져 있는 퍼셉트론을 단층 퍼셉트론이라고 해요.

인공 신경망이 학습이 진행되면 하얀색 직선이 생기면서 그 직선을 경계로 주황색 영역과 파란색 영역으로 나뉘는 것을 볼 수 있을거에요.  
학습이 진행됨에 따라서 OUTPUT 아래의 Test loss 값이 점점 작아지는 것을 볼 수 있을거에요.  
Test loss 값은 현재의 인공 신경망이 데이터를 얼마나 잘 분류 했는지를 나타내는 지표에요. 이 값이 작을수록 잘 분류한 거에요.

<h2 style='font-weight: bold'>4. 단층 퍼셉트론의 한계점은 무엇일까?</h2>
</div>

실생활에서는 단층 퍼셉트론으로 해결할 수 없는 문제도 존재해요.  
분포형태 1을 볼까요? 분포형태 1은 분포형태 3보다 조금 더 복잡한 분포를 가지고 있어요.  
이 분포를 분포형태 3처럼 직선 하나로 두 데이터를 나눌 수 있을까요??

정답부터 말하자면 이 데이터 분포는 하나의 직선으로 나눌 수 없어요.  
인공신경망을 학습시키면 학습 횟수가 늘어나도 Test loss가 줄어들지 않는 것을 확인할 수 있어요.  
이처럼 단층 퍼셉트론은 복잡한 데이터, 특히 비선형 데이터(직선으로 나눌 수 없는 데이터)를 잘 학습하지 못한다는 문제가 있어요.  
그러면 이러한 데이터는 어떻게 학습할 수 있을까요?

이를 위해서 2가지 방법을 사용할 수 있어요.  
첫 번째로 비선형인 입력을 주는 방법이 있어요. x1,x2 대신 x1^2과 x2^2을 입력으로 줘볼까요?

[x1^2, x2^2 사용했을 때]

위의 그림처럼 얼마 학습시키지 않아도 Test loss가 0에 가까워지고 데이터가 잘 나뉘어 진 것을 볼 수 있어요.  
두 번째는 다층 퍼셉트론을 사용하는 방법이 있답니다.  
다층 퍼셉트론에 대해서는 뒤에서 좀 더 자세하게 알아보도록 해요.

<h2 style='font-weight: bold'>5. 다층 퍼셉트론의 구조를 이해해보자.</h2>
</div>

다층 퍼셉트론은 분포형태 1이나 2와 같이 하나의 직선으로 분류할 수 없는 데이터를 분류하기 위해서 사용할 수 있어요.  
다층 퍼셉트론은 단층 퍼셉트론의 입력층과 출력층 사이에 새롭게 층을 추가하는 방법으로 만들 수 있어요.  
새롭게 추가된 층들을 은닉층(Hidden Layer)라고 해요.  
은닉층을 몇개 두어야 할 지, 각각의 은닉층에 몇 개의 퍼셉트론을 설정할 지는 인공 신경망을 만드는 여러분의 선택에 달려 있어요.  
하나 주의해야 할 점은 은닉층이 많고, 퍼셉트론이 많을수록 항상 좋은 결과를 내는 것은 아니라는 거에요.  
은닉층이 늘어나고 퍼셉트론이 많아지면 대체로 학습 시키는 데 걸리는 시간이 늘어난다는 단점도 있어요.

이후 학습을 시작하면 학습 횟수에 따라 두 데이터의 영역이 계속해서 변하는 것을 확인할 수 있을거에요.  
그리고 학습 횟수가 충분히 늘어나면 비선형 입력을 사용했을 때 처럼 데이터가 잘 분류되는 것을 볼 수 있어요.  
이처럼 복잡한 데이터를 분류하는 문제를 풀기 위해서 은닉층을 적절히 추가해 다층 퍼셉트론을 사용하는 방법을 사용할 수 있답니다.
